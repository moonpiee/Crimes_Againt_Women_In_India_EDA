```python
df.rename(columns={'Unnamed: 0':'S.no', 'STATE/UT':'State/UT', 'DISTRICT':'District'},inplace=True)
df.head(10)
```

```python
dumm_df=pd.DataFrame(df['State/UT'].str.replace(" ",""))
dumm_df
dumm_df=pd.DataFrame(dumm_df['State/UT'].str.lower())
# cmp_df=cmp_df['State/UT'].unique()
dumm_df.values
df['State/UT']=dumm_df['State/UT'].values
# df.replace(df['State/UT'],dumm_df['State/UT'], inplace=True)
df
```

```python
df['District'].replace("Total District(s)", "TOTAL", inplace=True)
df['District'][df['District'].str.contains("TOTAL")].unique()

# df['District'][df['District'].astype(str).str.contains("TOTAL")]
```

```python
df['District'].replace("DELHI UT TOTAL", "TOTAL", inplace=True)
df['District'].replace("ZZ TOTAL", "TOTAL", inplace=True)
df['District'][df['District'].astype(str).str.contains("TOTAL")].count()
```

```python
df=df.where(df['District']!="TOTAL").dropna()
```
